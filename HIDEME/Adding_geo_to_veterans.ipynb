```python
import pandas as pd
```

```python
file = 'Resources/sorted_veterans_data.csv'
veterans_df = pd.read_csv(file)
veterans_df.head(5)
```

```python
file_2 = 'Resources/statelatlong.csv'
state_df = pd.read_csv(file_2)
state_df.head(5)
```

```python
# combine state_df to veterans_df to incorporate lat and long
combined_df = veterans_df.merge(state_df, left_on='State', right_on='State', right_index=False)
```

```python
combined_df.head(5)
```

```python
cleaned_df = combined_df.drop(['Unnamed: 0', 'City'], axis=1)
cleaned_df.head(5)
```

```python
sorted_df = cleaned_df[["State","County","Latitude","Longitude","TotalVets","VeteranPopulationPct",
                       "MaleVetsPct","FemaleVetsPct","WhiteVetsPct","BlackVetsPct","HispanicVetsPct",
                       "OtherRaceVetsPct","LessThanHSVetsPct","HighSchOnlyVetsPct","SomeCollegeVetsPct",
                       "CollegeDegreeVetsPct","EmployeedVetsPct","UnemployeedVetsPct"]]
sorted_df.head(5)
```

```python
sorted_df.isnull().sum()
```

```python
sorted_df.to_json('Resources/veterans_loc.json')
```

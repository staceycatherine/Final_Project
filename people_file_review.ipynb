```python
import pandas as pd
import numpy as np
```

```python
file = 'Resources/People.csv'
people_df = pd.read_csv(file)
people_df.head()
```

```python
people_df.columns
```

```python
reduced_people_df = people_df[['State','County','TotalPopEst2015','TotalPopEst2016','TotalPopEst2017',
                              'TotalPopEst2018','TotalPopEst2019']]
reduced_people_df.head()
```

```python
reduced_people_df.fillna(0)
```

```python
# adding column for average population for years 2015 to 2019 to match veteran data for same timespan
reduced_people_df['AveragePop2015to2019'] = reduced_people_df.mean(axis=1)
reduced_people_df.head()
```

```python
# dropping row US data to eliminat outlier
reduced_people_df = reduced_people_df.drop(index=0, axis=0)
reduced_people_df.head()
```

```python
reduced_people_df.to_csv('Resources/reduced_people_data.csv',encoding='utf-8')
```

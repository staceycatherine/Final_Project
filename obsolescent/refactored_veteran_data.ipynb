```python
import pandas as pd
```

```python
file = 'Resources/veterans_pop_loc.csv'
df = pd.read_csv(file)
df.head()
```

```python
df.columns
```

```python
df_1 = df.drop('Unnamed: 0', axis=1)
df_1.head()
```

```python
df_1['VeteranPopulationPct'] = df_1['VeteranPopulationPct'].div(100)
df_1['MaleVetsPct'] = df_1['MaleVetsPct'].div(100)
df_1['FemaleVetsPct'] = df_1['FemaleVetsPct'].div(100)
df_1['WhiteVetsPct'] = df_1['WhiteVetsPct'].div(100)
df_1['BlackVetsPct'] = df_1['BlackVetsPct'].div(100)
df_1['HispanicVetsPct'] = df_1['HispanicVetsPct'].div(100)
df_1['OtherRaceVetsPct'] = df_1['OtherRaceVetsPct'].div(100)
df_1['LessThanHSVetsPct'] = df_1['LessThanHSVetsPct'].div(100)
df_1['HighSchOnlyVetsPct'] = df_1['HighSchOnlyVetsPct'].div(100)
df_1['SomeCollegeVetsPct'] = df_1['SomeCollegeVetsPct'].div(100)
df_1['CollegeDegreeVetsPct'] = df_1['CollegeDegreeVetsPct'].div(100)
df_1['EmployeedVetsPct'] = df_1['EmployeedVetsPct'].div(100)
df_1['UnemployeedVetsPct'] = df_1['UnemployeedVetsPct'].div(100)
df_1.head()
```

```python
df_1['VetPop'] = df_1['TotalVets'] * df_1['VeteranPopulationPct']
df_1['MaleVets'] = df_1['TotalVets'] * df_1['MaleVetsPct']
df_1['FemaleVets'] = df_1['TotalVets'] * df_1['FemaleVetsPct']
df_1['WhiteVets'] = df_1['TotalVets'] * df_1['WhiteVetsPct']
df_1['BlackVets'] = df_1['TotalVets'] * df_1['BlackVetsPct']
df_1['HispanicVets'] = df_1['TotalVets'] * df_1['HispanicVetsPct']
df_1['OtherRaceVets'] = df_1['TotalVets'] * df_1['OtherRaceVetsPct']
df_1['LessThanHSVets'] = df_1['TotalVets'] * df_1['LessThanHSVetsPct']
df_1['HighSchOnlyVets'] = df_1['TotalVets'] * df_1['HighSchOnlyVetsPct']
df_1['SomeCollegeVets'] = df_1['TotalVets'] * df_1['SomeCollegeVetsPct']
df_1['CollegeDegreeVets'] = df_1['TotalVets'] * df_1['CollegeDegreeVetsPct']
df_1['EmployeedVets'] = df_1['TotalVets'] * df_1['EmployeedVetsPct']
df_1['UnemployeedVets'] = df_1['TotalVets'] * df_1['UnemployeedVetsPct']
df_1.head()
```

```python
df_2 = df_1[['State','County','Latitude','Longitude','TotalVets','MaleVets',
            'FemaleVets','WhiteVets','BlackVets','HispanicVets','OtherRaceVets','LessThanHSVets',
            'HighSchOnlyVets','SomeCollegeVets','CollegeDegreeVets','EmployeedVets','UnemployeedVets']]
df_2.head()
```

```python
import numpy as np
df_2.dtypes
```

```python
df_2['TotalVets'] = df_2['TotalVets'].fillna(-1)
df_2['TotalVets'] = df_2['TotalVets'].astype(int)
df_2['MaleVets'] = df_2['MaleVets'].fillna(-1)
df_2['MaleVets'] = df_2['MaleVets'].astype(int)
df_2['FemaleVets'] = df_2['FemaleVets'].fillna(-1)
df_2['FemaleVets'] = df_2['FemaleVets'].astype(int)
df_2['WhiteVets'] = df_2['WhiteVets'].fillna(-1)
df_2['WhiteVets'] = df_2['WhiteVets'].astype(int)
df_2['BlackVets'] = df_2['BlackVets'].fillna(-1)
df_2['BlackVets'] = df_2['BlackVets'].astype(int)
df_2['HispanicVets'] = df_2['HispanicVets'].fillna(-1)
df_2['HispanicVets'] = df_2['HispanicVets'].astype(int)
df_2['OtherRaceVets'] = df_2['OtherRaceVets'].fillna(-1)
df_2['OtherRaceVets'] = df_2['OtherRaceVets'].astype(int)
df_2['LessThanHSVets'] = df_2['LessThanHSVets'].fillna(-1)
df_2['LessThanHSVets'] = df_2['LessThanHSVets'].astype(int)
df_2['HighSchOnlyVets'] = df_2['HighSchOnlyVets'].fillna(-1)
df_2['HighSchOnlyVets'] = df_2['HighSchOnlyVets'].astype(int)
df_2['SomeCollegeVets'] = df_2['SomeCollegeVets'].fillna(-1)
df_2['SomeCollegeVets'] = df_2['SomeCollegeVets'].astype(int)
df_2['CollegeDegreeVets'] = df_2['CollegeDegreeVets'].fillna(-1)
df_2['CollegeDegreeVets'] = df_2['CollegeDegreeVets'].astype(int)
df_2['EmployeedVets'] = df_2['EmployeedVets'].fillna(-1)
df_2['EmployeedVets'] = df_2['EmployeedVets'].astype(int)
df_2['UnemployeedVets'] = df_2['UnemployeedVets'].fillna(-1)
df_2['UnemployeedVets'] = df_2['UnemployeedVets'].astype(int)
df_2.head()
```

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
file = 'Resources/reduced_people_data.csv'
people_df = pd.read_csv(file)
people_df.head(5)
```

```python
people_df.dtypes
```

```python
file_2 = 'Resources/statelatlong.csv'
state_df = pd.read_csv(file_2)
state_df.head(5)
```

```python
state_df.dtypes
```

```python
# combine state_df to people_df to incorporate lat and long
combined_df = people_df.merge(state_df, left_on='State', right_on='State', right_index=False)
```

```python
combined_df.head(5)
```

```python
# drop unnecessary columns
cleaned_combined_df = combined_df.drop(['Unnamed: 0','TotalPopEst2015','TotalPopEst2016',
                                        'TotalPopEst2017','TotalPopEst2018','TotalPopEst2019','City'], axis=1)
cleaned_combined_df.head(5)
```

```python
sorted_df = cleaned_combined_df[["State","County","Latitude","Longitude","AveragePop2015to2019"]]
sorted_df.head(5)
```

```python
sorted_df.to_csv('Resources/people_pop_loc.csv', encoding='utf-8')
```
